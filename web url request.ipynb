```python
# Python3 program for a word frequency counter after crawling a web-page 
import requests 
from bs4 import BeautifulSoup 
import operator 
from collections import Counter 
  
'''Function defining the web-crawler/core 
spider, which will fetch information from 
a given website, and push the contents to 
the second  function clean_wordlist()'''
def start(url): 
  
    # empty list to store the contents of  
    # the website fetched from our web-crawler 
    wordlist = [] 
    source_code = requests.get(url).text 
  
    # BeautifulSoup object which will 
    # ping the requested url for data 
    soup = BeautifulSoup(source_code, 'html.parser') 
  
    # Text in given web-page is stored under 
    # the <div> tags with class <entry-content> 
    for each_text in soup.findAll('div', {'class':'entry-content'}): 
        content = each_text.text 
  
        # use split() to break the sentence into  
        # words and convert them into lowercase  
        words = content.lower().split() 
          
        for each_word in words: 
            wordlist.append(each_word) 
            
        clean_wordlist(wordlist) 
  
# Function removes any unwanted symbols 
def clean_wordlist(wordlist): 
      
    clean_list =[] 
    for word in wordlist: 
        symbols = '!@#$%^&*()_-+={[}]|\;:"<>?/., '
          
        for i in range (0, len(symbols)): 
            word = word.replace(symbols[i], '') 
              
        if len(word) > 0: 
            clean_list.append(word) 
            
    create_dictionary(clean_list) 
  
# Creates a dictionary conatining each word's  
# count and top_20 ocuuring words 
def create_dictionary(clean_list): 
    word_count = {} 
      
    for word in clean_list: 
        if word in word_count: 
            word_count[word] += 1
        else: 
            word_count[word] = 1
              
    ''' To get count of each word in 
        the crawled page --> 
          
    # operator.itemgetter() takes one  
    # parameter either 1(denotes keys) 
    # or 0 (denotes corresponding values) 
      
    for key, value in sorted(word_count.items(), 
                    key = operator.itemgetter(1)): 
        print ("% s : % s " % (key, value)) 
          
    <-- '''
  
      
    c = Counter(word_count) 
      
    # returns the most occuring elements 
    top = c.most_common(10) 
    print(top) 
  
# Driver code 
if __name__ == '__main__': 
    start("https://www.geeksforgeeks.org/programming-language-choose/") 
```

```python
import re
import urllib.request
#import operator
from collections import Counter

# Download wiki page
url = "https://www.national-lottery.co.uk/results/euromillions/draw-history"
html = urllib.request.urlopen(url).read()

# Find all mentioned years in the 20th or 21st century
# to create a non-capturing group, use the syntax (?:pattern)
regex = r'(\d{2} - \d{2} - \d{2} - \d{2} - \d{2})'     
fives = re.findall(regex, str(html))
#print( re.match(regex, str(html)).group())
#fives = fives[1::]

for i, strnum in enumerate(fives):
    nums = strnum.split(" - ")             # split returns a list
    intnum = [int(j) for j in nums]   # Convert all strings in a list to int
    fives[i] = intnum

regex2 = r'(  \d{2} - \d{2})\\n'
twos = re.findall(regex2, str(html))

for i, strnum in enumerate(twos):
    nums = strnum.split(" - ")
    intnum = [int(j) for j in nums]
    twos[i] = intnum
    
win_nums = list(zip(fives, twos))
# print(win_nums[0:5])

#get first 5 win numbers
l = [n for outer in fives[0:5] for n in outer]
# print('first five win numbers without bonus\n ', sorted(l))

xl = list(set(l))   # remove dupes
al = list(range(1,51))  # 1-50

common_num = [al_1 for al_1 in al for xl_1 in xl if al_1 == xl_1]
# print('commom_nums from last five wins \n', common_num)

unique_num =  [a for a in al if a not in xl]
print('unique nums\n', unique_num)


import random
#print(random.choice(yy))
total = []

for i in range(999):
    guess = []
    while len(guess) < 5:
        x = random.choice(unique_num)
        if x not in guess:
            guess.append(x)      
    
    guess.sort()
    #print(guess)
    
    total += guess

## Counter returns dictionary sorted by value
d = Counter(total)
l = Counter.most_common(d)
print(l[0:7])

final_guess = [y[0] for y in l[0:10]]
final_bonus = [y[0] for y in l if y[0] < 13]

# sorted     is a built-in function that is available in a standard installation of Python.
# print('final guess = ', sorted(final_guess))  # a new, sorted list
print('final guess = ', final_guess) 
print('final bonus = ', final_bonus[0:5])

# First, .sort() is a method of the list class and can only be used with lists.
# 2nd,  .sort() returns None and modifies the values in place.
```

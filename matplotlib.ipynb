```python
# << Vectorization ??? >>
# Many CPUs have "vector" or "SIMD" instruction sets which apply the same operation simultaneously 
# to two, four, or more pieces of data. 
# Modern x86 chips have the SSE instructions, many PPC chips have the "Altivec" instructions, 
# and even some ARM chips have a vector instruction set, called NEON.

# "Vectorization" (simplified) is the process of rewriting a loop so that instead of processing 
# a single element of an array N times, it processes (say) 4 elements of the array simultaneously N/4 times.

# (I chose 4 because it's what modern hardware is most likely to directly support; 
# the term "vectorization" is also used to describe a higher level software transformation 
# where you might just abstract away the loop altogether and just describe operating on arrays 
# instead of the elements that comprise them)




import numpy as np
# If matplotlib were limited to working with lists, it would be fairly useless for numeric processing. 
# Generally, you will use numpy arrays. In fact, all sequences are converted to numpy arrays internally.

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import csv
from collections import Counter

#print(plt.style.available)
plt.style.use('fivethirtyeight')

"""
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()  # empty counter
    
    #row = next(csv_reader)
    #print(row)
    #print(row['LanguagesWorkedWith'].split(';'))
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
"""

#### replaced the open file with pandas

data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()  # empty counter

for response in lang_responses:
        language_counter.update(response.split(';'))
        
####

#print(language_counter)
#print(language_counter.most_common(15))
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
#print(languages)
#print(popularity)

#plt.bar(languages, popularity)

languages.reverse()
popularity.reverse()
plt.barh(languages, popularity)

plt.title('Most Popular Languages')
plt.xlabel('Number of People Who Use')
#plt.ylabel('Programming Languages')


plt.tight_layout()

plt.show()
```

```python
# Filling Area on Line Plots

from matplotlib import pyplot as plt
import pandas as pd

#print(plt.style.available)
plt.style.use('fivethirtyeight')

data = pd.read_csv('data2.csv')
#print(data)
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

#plt.fill_between(ages, py_salaries, overall_median,
#                where=(py_salaries > overall_median),
plt.fill_between(ages, py_salaries, dev_salaries,
                where=(py_salaries > dev_salaries),
                interpolate=True, alpha=0.25, label='Above Avg')

#plt.fill_between(ages, py_salaries, overall_median,
#                where=(py_salaries <= overall_median),
plt.fill_between(ages, py_salaries, dev_salaries,
                where=(py_salaries <= dev_salaries),
                interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
#plt.ylabel('Medisan Salary (USD)')


plt.tight_layout()

plt.show()
```

```python
# pie chart is best for less than 5 items

from matplotlib import pyplot as plt


#print(plt.style.available)
plt.style.use('fivethirtyeight')

#slices = [120, 80, 30, 20]
#labels = ['First', 'Second', 'Extra1', 'Extra2']
colors = ['blue', 'red', 'yellow', 'green', 'pink']

slices = [27, 24, 20, 20, 15]
labels = ['JavaScript', 'HTML/CSS', 'Python', 'SQL', 'Bash/Shell/PowerShell']
explode = [0,0,0,0.1,0]


plt.pie(slices, labels=labels, colors=colors, explode=explode, shadow=True, 
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})


plt.title('My Pie Chart')

plt.tight_layout()

plt.show()
```
